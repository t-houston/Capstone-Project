```python
# Import libaries
import requests
import numpy as np
import pandas as pd
```

```python
#pulls in csv's from data folder and turns them into dataframes
df09 = pd.read_csv('./Data/sportsref_2009.csv')
df10 = pd.read_csv('./Data/sportsref_2010.csv')
df11 = pd.read_csv('./Data/sportsref_2011.csv')
df12 = pd.read_csv('./Data/sportsref_2012.csv')
df13 = pd.read_csv('./Data/sportsref_2013.csv')
df14 = pd.read_csv('./Data/sportsref_2014.csv')
df15 = pd.read_csv('./Data/sportsref_2015.csv')
df16 = pd.read_csv('./Data/sportsref_2016.csv')
df17 = pd.read_csv('./Data/sportsref_2017.csv')
df18 = pd.read_csv('./Data/sportsref_2018.csv')
df19 = pd.read_csv('./Data/sportsref_2019.csv')
```

```python
#Adding the year of the player's recorded stats as 'YEAR' column to each dataframe
df09['YEAR'] = 2009
df10['YEAR'] = 2010
df11['YEAR'] = 2011
df12['YEAR'] = 2012
df13['YEAR'] = 2013
df14['YEAR'] = 2014
df15['YEAR'] = 2015
df16['YEAR'] = 2016
df17['YEAR'] = 2017
df18['YEAR'] = 2018
df19['YEAR'] = 2019
```

```python
#displays first five rows of the df09 dataframe to check that 'YEAR' column was added
df09.head()
```

```python
#combines all the dataframes into one big dataframe
df = pd.concat([df09,df10,df11,df12,df13,df14,df15,df16,df17,df18,df19])

#shows us number of rows/columns in the new combined df
df.shape
```
