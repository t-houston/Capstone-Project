# Linear Regression Modeling

most linear to 'num games' according to graphs:['MP', 'PER','OWS', 'DWS', 'WS', 'WS_40'],


```python
#create features variable
features = ['MP','DWS', 'WS_40', 'WS', 'PER', 'TOV%']
# '3PAr', 
#        'ORB%', 'TRB%', 'AST%', 'STL%', 'BLK%', 'USG%', 'OWS', 'DWS',
#        'WS', 'WS_40', 'YEAR']

#split data into x and y variables
X = df[features]
y = df['NUM GAMES']
# y = np.log(df['NUM GAMES'])

#run train,test,split on variables
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.7, random_state=42) 

# instantiate Standard Scaler and transform X train and test
ss = StandardScaler()
X_train_sc = ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)

```

```python
# Instantiate model
linmod = LinearRegression()

# Fit the regression model to the training data.
linmod.fit(X, y)

# make our predictions on training data
lrpredict = linmod.predict(X)
residuals = y - lrpredict
print(f'Mean Residual Score: {residuals.mean()}')
print()
```

```python
def r2_adj(y, lrpredict):  
    r2 = r2_score(y, lrpredict)
    n = y.shape[0]
    k = X.shape[1]
    r2adj = 1 - ((1 - r2) * (n - 1)) / (n - k - 1)
    return r2adj

def reg_metrics(X,y):
    print(f' RMSE: {np.sqrt(mean_squared_error(y, lrpredict))}') ##how many games am I off on avg
    print(f' R2: {linmod.score(X, y)}')
    print(f' Adjusted R-Squared: {r2_adj(y, lrpredict)}')
    print(f' MSE: {metrics.mean_squared_error(y, lrpredict)}')
    print(f' Median Absolute Error: {metrics.median_absolute_error(y, lrpredict)}')
    print(f' Mean Absolute Error: {metrics.mean_absolute_error(y, lrpredict)}') 
    
reg_metrics(X,y)

#evaluate the coeffs -these contribute most to prediction
# Print out intercept and coefficients.
print(f'Linear Regression Coefficient: {linmod.coef_}')
print(f'Linear Regression Intercept: {linmod.intercept_}')
```

```python
# Instantiate models
linmod = LinearRegression()
lasso = LassoCV(n_alphas=200)
ridge = RidgeCV(alphas=np.linspace(.1, 10, 100))

# Fit our regression models to the training data.
linmod.fit(X_train,y_train)
lasso.fit(X_train,y_train)
ridge.fit(X_train,y_train)

#g Prints the crossvalidation scores
linmod_scores = cross_val_score(linmod, X_train, y_train, cv=3)
print(f'Linear Regression cross_val_mean: {linmod_scores.mean()}')
lasso_scores = cross_val_score(lasso, X_train, y_train, cv=3)
print(f'Lasso Regression cross_val_mean: {lasso_scores.mean()}')
ridge_scores = cross_val_score(ridge, X_train, y_train, cv=3)
print(f'Ridge Regression cross_val_mean: {ridge_scores.mean()}')
print()


# Prints out intercept and coefficients.
print(f'Linear Regression Intercept: {linmod.intercept_}')
print(f'Lasso Intercept: {lasso.intercept_}')
print(f'Ridge Intercept: {ridge.intercept_}')
print()

print(f'Linear Regression Coefficient: {linmod.coef_}')
print(f'Lasso Coefficient: {lasso.coef_}')
print(f'Ridge Coefficient: {ridge.coef_}')
print()


# Make predictions on training data
lrpred_train = linmod.predict(X_train)
lassopred_train = lasso.predict(X_train)
ridgepred_train = ridge.predict(X_train)


# Make our predictions on testing data
lrpred_test = linmod.predict(X_test)
lassopred_test = lasso.predict(X_test)
ridgepred_test = ridge.predict(X_test)

# Calculates residuals for testing data with predictions made above
lrresiduals = y_test - lrpred_test
lassoresiduals = y_test - lassopred_test
ridgeresiduals = y_test - ridgepred_test

#calculates model's R2 scores on train data
print(f'Linear Regression Train R2 Score: {linmod.score(X_train, y_train)}')
print(f'Lasso Train R2 Score: {lasso.score(X_train, y_train)}')
print(f'Ridge Train R2 Score: {ridge.score(X_train, y_train)}')
print()


#calculates model's R2 scores on test data
print(f'Linear Regression Test R2 Score: {linmod.score(X_test, y_test)}')
print(f'Lasso Test R2 Score: {lasso.score(X_test, y_test)}')
print(f'Ridge Test R2 Score: {ridge.score(X_test, y_test)}')
print()


# 1 value increase in x + coeff
# as turnover rate increases by 1% num games that blah blah decreases by coeff - significant? not? why is it in there if not? 
# ols to get p values? tell me which ones to inclue in model? why is it important?
```

```python
# Gets MSE for training 
print(f'Linear Regression Train MSE: {metrics.mean_squared_error(y_train, lrpred_train)}')
print(f'Lasso Train MSE: {metrics.mean_squared_error(y_train, lassopred_train)}')
print(f'Ridge Train MSE: {metrics.mean_squared_error(y_train, ridgepred_train)}')
print()

# Gets MSE for testing 
print(f'Linear Regression Test MSE: {metrics.mean_squared_error(y_test, lrpred_test)}')
print(f'Lasso Test MSE: {metrics.mean_squared_error(y_test, lassopred_test)}')
print(f'Ridge Test MSE: {metrics.mean_squared_error(y_test, ridgepred_test)}')
print()

# Gets the RMSE for training
print(f'Linear Regression Train RMSE: {np.sqrt(metrics.mean_squared_error(y_train, lrpred_train))}')
print(f'Lasso Train RMSE: {np.sqrt(metrics.mean_squared_error(y_train, lassopred_train))}')
print(f'Ridge Train RMSE: {np.sqrt(metrics.mean_squared_error(y_train, ridgepred_train))}')
print()

                                
# Gets the RMSE for testing
print(f'Linear Regression Test RMSE: {np.sqrt(metrics.mean_squared_error(y_test, lassopred_test))}')
print(f'Lasso Test RMSE: {np.sqrt(metrics.mean_squared_error(y_test, lassopred_test))}')
print(f'Ridge Test RMSE: {np.sqrt(metrics.mean_squared_error(y_test, ridgepred_test))}')
print()
```

# Shrubs - RF

```python
#create features variable
features = ['MP','DWS', 'WS_40', 'WS', 'PER']
# '3PAr', 
#        'ORB%', 'TRB%', 'AST%', 'STL%', 'BLK%', 'USG%', 'OWS', 'DWS',
#        'WS', 'WS_40', 'YEAR']
#split data into x and y variables
X = df[features]
y = df['NUM GAMES']

#run train,test,split on variables
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    train_size=.7, 
    random_state=42) 

# instantiate Standard Scaler and transform X train and test
ss = StandardScaler()
X_train_sc = ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)

#instantiate and fit our Random Forest Regression model
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)

# Print 
print(f'Random Forest Score on Training (No GS): {rfr.score(X_train, y_train)}')
print(f'Random Forest Score on Testing (No GS): {rfr.score(X_test, y_test)}')
```

```python
# Set Random Forest Regressor parameters
rfr_params = {
    'n_estimators': [100, 110, 120],
    'max_depth': [None, 4, 5, 6, 7],
}

# Instantiate our GridSearchCV object.
rfr_gs = GridSearchCV(rfr, # What is the model we want to fit?
                      rfr_params, # What is the dictionary of hyperparameters?
                      cv=5, # What number of folds in CV will we use?
                      verbose=1
                      )

# Fit our Gridsearch on to our Random Forest Regressor 
rfr_gs.fit(X_train, y_train)
```

```python
# Print out the set of hyperparameters that achieved the best score.
print(f'RFR Best Param: {rfr_gs.best_params_}')

# Prints the best accuracy scores for both the training and testing data.
print(f'Random Forest GS Score on Training: {rfr_gs.score(X_train, y_train)}')
print(f'RFR Best Test GS Score: {rfr_gs.score(X_test, y_test)}')
```

```python
#create function for returning RMSE score on both training and testing data
def rmse_score(model, X_train, X_test, y_train, y_test):
    mse_train = mean_squared_error(y_true = y_train,
                                  y_pred = model.predict(X_train))
    mse_test = mean_squared_error(y_true = y_test,
                                  y_pred = model.predict(X_test))
    rmse_train = mse_train ** 0.5
    rmse_test = mse_test ** 0.5
    
    print("The Training RMSE for " + str(model) + " is: " + str(rmse_train))
    print()
    print("The Testing RMSE for " + str(model) + " is: " + str(rmse_test))
    return (rmse_train, rmse_test)

#use function on random forest regeressor model
rmse_score(rfr, X_train, X_test, y_train, y_test)
print()
rmse_score(rfr_gs, X_train, X_test, y_train, y_test)
```
